##### Импортируем данные

```python
def data_from_g_sheets(spreadsheet_id):
    file_name = 'https://docs.google.com/spreadsheets/d/{}/export?format=csv'.format(spreadsheet_id)
    r = requests.get(file_name)
    return pd.read_csv(BytesIO(r.content))
```

```python
try:
    df_planes = pd.read_csv('/datasets/query_1.csv')
    df_cities = pd.read_csv('/datasets/query_3.csv')
except:
    df_planes = data_from_g_sheets('18IDip25SIw9nqUsItr6iq4vNjy0IizW5BgcDa3s-Xic')
    df_cities = data_from_g_sheets('1Hom-Hknwo5FU4HT0Fyf9fuu9kNtGpzrj9B2mBFhwQqc')
```

## II. Исследовательский анализ <a id="analysis"></a>

##### Посмотрим топ-10 городов по среднедневному количеству рейсов в августе 2018 года

```python
df_cities.sort_values('average_flights', ascending=False).head(10)
```

##### Вывод: 
Со значительным отрывом от остальных идет Москва - самое популярное направление из нашей выборки

##### Построим график количество рейсов по каждой модели самолета из нашей выборки

```python
df_planes_for_plot = df_planes.sort_values('flights_amount', ascending=False)
```

```python
x = df_planes_for_plot['model']
y = df_planes_for_plot['flights_amount']

fig, ax = plt.subplots()

ax.bar(x, y, width = 0.6)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

fig.set_figwidth(18)
fig.set_figheight(6)

plt.title('Количество рейсов по каждой модели самолетов в сентябре 2019')
plt.xlabel('Модели самолетов')
plt.ylabel('Количество рейсов')
plt.show()
```

##### Вывод: 
- Больше всего в сентбяре было рейсов у самолетов моделей: Cessna 208 Caravan, Bombardier CRJ-200, Sukhoi Superjet-100.
В этом распределении есть зависимость от размера модели самолета. Чем больше самолет, тем реже он летает. Например, Cessna 208 Caravan - это маленький самолетик с 9-ю пассажирскими местами. Из открытых источников мы выяснили, что в России этих самолетов больше всего у Авиационной компания „ПАНХ“ - российской авиационной компании, базирующаяся в Улан-Удэ и выполняющей чартерные и регулярные пассажирские и грузовые перевозки. Веротяно, и рейсы у этих самолетов по Забайкальскому краю, там где плохо с дорогами. Дальность полетов этих самолетов всего 1200 км. Летают недалеко, но часто. В другой стороне графика у нас Boeing 777-300, который летает " редко, но метко" :-) Его вместимость 500 пассажиров, но и дальность палета у него 10500 км.


- Наш топ-3 самых летающих самолетов - это модели с вместимостью до 100 пассажиров и ближних рейсов;
- Реже всего летают модели для дальних рейсов вместимостью от 150 до 500 человек.



##### Построим график среднего количества рейсов в день в разбивке по городам

```python
df_cities_for_plot = df_cities.sort_values('average_flights').reset_index()
```

```python
x = df_cities_for_plot['city']
y = df_cities_for_plot['average_flights']

fig, ax = plt.subplots()

ax.barh(x, y)

fig.set_figwidth(10)
fig.set_figheight(25)

plt.title('Среднее количество рейсов в день в августе 2018')
plt.xlabel('Среднее кол-во рейсов в день')
#plt.ylabel('Города')
plt.show()
```

##### Вывод:
Больше всего рейсов приходится на Москву. Столица Родины как-никак. Так же, мы думаем, что это связно с тем, что из многих регинов нельзя улететь по всем международным направлениям, а из Москвы можно. Соотвественно люди летят со всей России в Москву, а уже потом зарубеж.

##### Построим график среднего количества рейсов в день в разбивке по городам ТОП-10

```python
df_cities_for_plot_10 = df_cities.sort_values('average_flights').tail(10).reset_index()
```

```python
df_cities_for_plot_10
```

```python
x = df_cities_for_plot_10['city']
y = df_cities_for_plot_10['average_flights']

fig, ax = plt.subplots()

ax.barh(x, y)

fig.set_figwidth(5)
fig.set_figheight(6)

plt.title('Среднее количество рейсов в день в августе 2018')
plt.xlabel('Среднее кол-во рейсов в день')
#plt.ylabel('Города')
plt.show()
```

##### Вывод:
Топ-10 по количеству рейсов получился весьма очевидным. Как правило, люди летят в эти города, потому что это административные, культурные и экономические центры. Помимо этого, все города из Топ-10 - города миллионники, (кроме Сочи). Во всех этих городах есть международные аэропорты. Сочи попал в Топ, потому что это самый популярный курорт в России. В августе разгар сезона отпусков. Многие летят к морю.
