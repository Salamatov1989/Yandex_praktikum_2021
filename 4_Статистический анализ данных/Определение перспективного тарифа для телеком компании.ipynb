#### Посчитаем помесячную выручку от каждого пользователя: <a id="calc_2"></a>
- вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; 
- остаток умножим на значение из тарифного плана; 
- прибавим абонентскую плату, соответствующую тарифному плану.

```python
df_all_piv['minutes_overrun']  = df_all_piv['minutes'] - df_all_piv['minutes_included']
df_all_piv['messages_overrun'] = df_all_piv['messages'] - df_all_piv['messages_included']

# Мегабайты перевожу в гигабайты, убираю NaN-ы, чтобы использовать функцию округления ceil. с NaN-ами она не работает.
df_all_piv['gb_overrun'] = (df_all_piv['mb'] - df_all_piv['mb_per_month_included'])/1024
df_all_piv['gb_overrun'] = df_all_piv['gb_overrun'].fillna(0)
# Округялю вверх, чтобы потом корректно посчитать выручку сверхлимита.
df_all_piv['gb_overrun'] = df_all_piv['gb_overrun'].apply(lambda x: math.ceil(x))
```

```python
def get_revenue(row):
    revenue_by_min = 0
    revenue_by_messages = 0
    revenue_gb = 0
    if row['minutes_overrun'] > 0:
        revenue_by_min = row['minutes_overrun'] * row['rub_per_minute']
    if row['messages_overrun'] > 0:
        revenue_by_messages = row['messages_overrun'] * row['rub_per_message']
    if row['gb_overrun'] > 0:
        revenue_gb = (row['gb_overrun']) * row['rub_per_gb']
    return revenue_by_min + revenue_by_messages + revenue_gb
```

```python
df_all_piv['revenue_overrun'] = df_all_piv.apply(get_revenue, axis=1)
```

```python
display(df_all_piv.sample(10))
```

```python
df_all_piv['total_revenue'] = df_all_piv['rub_monthly_fee'] + df_all_piv['revenue_overrun']
```

## IV. Проверим гипотезы<a id="fouth"></a>

### Гипотеза №1:  <a id="hip_1"></a>
- Нулевая гипотеза: Cредняя выручка от пользователей тарифа «Ультра» такая же как и от абонентов на тарифе «Смарт»;
- Альтернативная гипотеза: Cредняя выручка от пользователей тарифов «Ультра» и «Смарт» различается;
- Уровень значимости: α == 0.05;
- Посчитаем p-value, используя ttest из библиотеки scipy для сравнения среди двух выборок;

- Если p-value будет меньше выбранного уровня значимости α == 0.05, то нулевую гипотезу о равенстве средней выручки двух разных тарифов отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

##### Для начала посмотрим на распределение

```python
smart = df_all_piv.query('tariff_name=="smart"')['total_revenue']
ultra = df_all_piv.query('tariff_name=="ultra"')['total_revenue']
```

```python
plt.figure(figsize=(8,8))

x = smart
y = ultra

bins = numpy.linspace(0, 5000, 20)
plt.hist(x, bins, alpha=0.6, label='smart')
plt.hist(y, bins, alpha=0.3, label='ultra')

plt.grid(True)
plt.title('Распределение выручки')
plt.legend(loc='upper right')
plt.xlabel('Выручка, руб')
plt.ylabel('Количество абонентов, шт')
plt.show()
```

##### Распеределение непохоже на нормальное, это скорее распеределение Пауссона. Значит нам стоит провести стат.тест  Левене, чтобы понять с каким параметром `equal_var`  считать p-value

```python
from scipy.stats import levene

stat, p = levene(smart, ultra)
p
```

```python
[np.var(x, ddof=1) for x in [smart, ultra]]
```

##### Малое значение p предполагает, что совокупности не имеют одинаковых дисперсий. Значит мы будем использововать `equal_var=False`

```python
from scipy import stats as st
import numpy as np

smart = df_all_piv.query('tariff_name=="smart"')['total_revenue']
ultra = df_all_piv.query('tariff_name=="ultra"')['total_revenue']

print('Средняя выручка от пользователей ТП Smart: ', smart.mean())  
print('Средняя выручка от пользователей ТП Ultra: ', ultra.mean())


alpha = .05

results = st.ttest_ind(
    smart, 
    ultra,
    equal_var = False)

print('p-значение:',results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

- Если p-value будет меньше выбранного уровня значимости α == 0.05, то нулевую гипотезу о равенстве средней выручки по Москве и регионам отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

##### Для начала посмотрим на распределение

```python
moscow = df_all_piv.query('city=="Москва"')['total_revenue']
region = df_all_piv.query('city!="Москва"')['total_revenue']
```

##### Распеределение непохоже на нормальное, это скорее распеределение Пауссона. Значит нам стоит провести стат.тест  Левене, чтобы понять с каким параметром `equal_var`  считать p-value.

```python
from scipy.stats import levene

stat, p = levene(moscow, region)
p
```

```python
print('Средняя выручка от пользователей из Москвы: ', moscow.mean())  
print('Средняя выручка от пользователей из Регионов: ', region.mean())


alpha = .05 

results = st.ttest_ind(
    moscow, 
    region,
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```
