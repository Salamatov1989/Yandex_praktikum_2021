# Стратегия взаимодействия с клиентами на основе аналитических данных сети фитнес-центров «Культурист-датасаентист»

## Описание проекта:

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.
Распространённая проблема фитнес-клубов и других сервисов — отток клиентов.
Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. 
Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Наша задача — провести анализ и подготовить план действий по удержанию клиентов.

    1) выделить целевые группы клиентов;

## Описание данных:

Путь к файлу: /datasets/gym_churn.csv
Набор данных включает следующие поля:

'Churn' — факт оттока в текущем месяце;

Текущие поля в датасете:
    
Данные клиента за предыдущий до проверки факта оттока месяц:
    
- 'gender' — пол;
- 'Near_Location' — проживание или работа в районе, где находится фитнес-центр;
- 'Partner' — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);

Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:

# III. Построим модель прогнозирования оттока клиентов <a id="reserch"></a>

###### Для начала удалим ненужные признаки 

```python
gym_churn_clear = gym_churn.drop(['Phone', 'gender', 'Contract_period', 'Avg_class_frequency_current_month'], axis=1)
```

```python
gym_churn_clear.head(5)
```

###### Разделим наши данные на признаки (матрица X) и целевую переменную (y)

```python
X = gym_churn_clear.drop(columns = ['Churn'])
y = gym_churn_clear['Churn']
```

###### Разобьем данные на обучающую и валидационную выборку функцией train_test_split()

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

###### Создадим объект класса StandardScaler и применим его к обучающей выборке

```python
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
```

###### Обучим модель на train-выборке двумя способами: логистической регрессией и случайным лесом

```python
models = [LogisticRegression(random_state=0) , RandomForestClassifier(random_state=0)]
```

```python
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy: {:.2f} Precision: {:.2f} Recall: {:.2f}'.format(accuracy_score(y_test, y_pred),
                                                                     precision_score(y_test, y_pred),
                                                                     recall_score(y_test, y_pred)))
```

###### Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке

```python
for i in models:
    print(i)
    make_prediction(i, X_train, y_train, X_test, y_test)
```

# IV. Сделаем кластеризацию клиентов <a id="claster"></a>

###### Стандартизируем данные

```python
scaler = StandardScaler()
x_sc = scaler.fit_transform(gym_churn_clear.drop(columns = ['Churn']))
print(x_sc[:10]) 
```

###### Построим матрицу расстояний (дендрограмму) функцией linkage()

```python
linked = linkage(x_sc, method = 'ward')
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.show() 
```

###### Дендограмма предлагает нам разбить нашу выборку на 3 кластера

###### Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов.  За число кластеров принмем n=5

```python
km = KMeans(n_clusters = 5, random_state=0) ### нужен random_state
labels = km.fit_predict(x_sc)
```

###### Сохраняем метки кластера в поле нашего датасета

```python
gym_churn_clear['cluster_km'] = labels
```

###### Посмотрим на средние значения признаков для кластеров

```python
display(gym_churn_clear.groupby('cluster_km').agg('mean'))
```

###### Построим распределения признаков для кластеров

```python
names = list(gym_churn_clear.columns)
names
```

```python
claster_0 = gym_churn_clear.loc[gym_churn_clear['cluster_km'] == 0]
claster_1 = gym_churn_clear.loc[gym_churn_clear['cluster_km'] == 1]
claster_2 = gym_churn_clear.loc[gym_churn_clear['cluster_km'] == 2]
claster_3 = gym_churn_clear.loc[gym_churn_clear['cluster_km'] == 3]
claster_4 = gym_churn_clear.loc[gym_churn_clear['cluster_km'] == 4]
```

```python
for name in names:
    plt.figure(figsize = (12,5))
    plt.hist(claster_0[name], bins=20, alpha=0.2, label='claster_0')
    plt.hist(claster_1[name], bins=20, alpha=0.6, label='claster_1')
    plt.hist(claster_2[name], bins=20, alpha=0.2, label='claster_2')
    plt.hist(claster_3[name], bins=20, alpha=0.2, label='claster_3')
    plt.hist(claster_4[name], bins=20, alpha=0.2, label='claster_4')
    plt.title(name)
    plt.legend(loc='upper right')
    plt.show()
```

```python
name = 'Lifetime'
min_value = min(gym_churn_clear[name])
max_value = max(gym_churn_clear[name])

plt.figure(figsize = (12,5))
plt.hist(claster_0[name], bins=20, alpha=0.2, label='claster_0', density=True, range=(min_value, max_value))
plt.hist(claster_1[name], bins=20, alpha=0.6, label='claster_1', density=True, range=(min_value, max_value))
plt.hist(claster_2[name], bins=20, alpha=0.2, label='claster_2', density=True, range=(min_value, max_value))
plt.hist(claster_3[name], bins=20, alpha=0.2, label='claster_3', density=True, range=(min_value, max_value))
plt.hist(claster_4[name], bins=20, alpha=0.2, label='claster_4', density=True, range=(min_value, max_value))
plt.title(name)
plt.legend(loc='upper right')
plt.show()
```
